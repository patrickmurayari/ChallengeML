```python
# En notebooks/limpieza_integracion.ipynb
import pandas as pd
```

```python
DATA_PATH = '../data/raw/'
df_recorridos = pd.read_csv(DATA_PATH + 'badata_ecobici_recorridos_realizados_2024.csv')
df_usuarios = pd.read_csv(DATA_PATH + 'usuarios_ecobici_2024.csv')
```

# --- 1. df_recorridos: Renombrar columnas a snake_case ----

```python
df_recorridos.head(10)
```

```python
df_usuarios.head(10)
```

```python
print("Columas originales de df_recorridos")
print(df_recorridos.columns.tolist())
```

```python
# Crea un diccionario de mapeo o usa una funcion para renombrar:
# Una forma rápida es definir una función para convertir el texto:
def to_snake_case(col_name):
    # Simplificado: convertir a minúsculas y reemplazar espacios/puntos/guiones por guiones bajos.
    return col_name.lower().replace(' ', '_').replace('.', '_').strip()

df_recorridos.columns = [to_snake_case(col) for col in df_recorridos.columns]
# Aquí deberás revisar la salida de print(df_recorridos.columns) para ver si necesitas ajustes manuales.
```

```python
print("\nColumnas en snake_case de df_recorridos:")
print(df_recorridos.columns.tolist())
```

```python
# --- 2. df_usuarios: Renombrar columnas a snake_case ---
print("\nColumnas originales de df_usuarios:")
print(df_usuarios.columns.tolist())
```

```python
df_usuarios.columns = [to_snake_case(col) for col in df_usuarios.columns]
print("\nColumnas en snake_case de df_usuarios:")
print(df_usuarios.columns.tolist())
```

# --- 3. Conversión de Tipos de Datos ---

```python
# Columnas de df_recorridos que deben ser datetime:
date_cols_recorridos = ['fecha_origen_recorrido', 'fecha_destino_recorrido']

for col in date_cols_recorridos:
    # 'errors='coerce' convierte valores inválidos (como los nulos) a NaT (Not a Time)
    df_recorridos[col] = pd.to_datetime(df_recorridos[col], errors='coerce')
```

```python
# Columnas de df_usuarios que deben ser datetime:
# Si el archivo de Usuarios tenía 'fecha_alta' y 'hora_alta' separadas, es mejor unirlas primero.
# Supongamos que df_usuarios ahora tiene 'fecha_alta' y 'hora_alta' en snake_case
df_usuarios['fecha_hora_alta'] = pd.to_datetime(
    df_usuarios['fecha_alta'] + ' ' + df_usuarios['hora_alta'], errors='coerce'
)
```

```python
# Puedes eliminar las columnas originales después de la unión si quieres
df_usuarios = df_usuarios.drop(columns=['fecha_alta', 'hora_alta'])
```

```python
print("\nTipos de datos de df_recorridos después de la conversión:")
print(df_recorridos[date_cols_recorridos].dtypes)
print("\nTipos de datos de df_usuarios después de la conversión:")
print(df_usuarios[['fecha_hora_alta']].dtypes)
```

# Filtrado, Preparacion e Integracion de Datos

```python
# En notebooks/limpieza_integracion.ipynb

# --- 4. Filtrado Temporal: Julio, Agosto y Septiembre de 2024 ---

# Definir las fechas de inicio y fin del periodo
# Es importante incluir el inicio del primer día y el final del último día
inicio_periodo = pd.to_datetime('2024-07-01')
fin_periodo = pd.to_datetime('2024-09-30 23:59:59')

# Crear la máscara de filtrado
# Se asegura que la fecha esté dentro del rango [inicio_periodo, fin_periodo]
df_filtrado = df_recorridos[
    (df_recorridos['fecha_origen_recorrido'] >= inicio_periodo) & 
    (df_recorridos['fecha_origen_recorrido'] <= fin_periodo)
].copy() # Usamos .copy() para evitar SettingWithCopyWarning

print(f"Total de viajes filtrados (Julio a Septiembre 2024): {len(df_filtrado):,}")
```

```python
# --- 5. Guardar el resultado intermedio ---
INTERIM_PATH = '../data/interim/'
df_filtrado.to_csv(INTERIM_PATH + 'recorridos_3meses.csv', index=False)

print(f"Guardado intermedio exitoso en {INTERIM_PATH}recorridos_3meses.csv")
```

```python
# --- 6. Realizar la Integración (JOIN) ---

# Revisa la columna de unión en df_usuarios si la renombraste (asumo que es 'id_usuario')
print("\nVerificación de claves de unión:")
print(f"Recorridos: 'id_usuario' tipo {df_filtrado['id_usuario'].dtype}")
print(f"Usuarios: 'id_usuario' tipo {df_usuarios['id_usuario'].dtype}")
```

```python
# Realizar el merge (join)
df_final = pd.merge(
    left=df_filtrado,
    right=df_usuarios,
    on='id_usuario',
    how='left' # Mantiene todos los viajes de df_filtrado
)

print(f"\nFilas después del merge: {len(df_final):,}")
print("Columnas después del merge (deberían ser más):")
print(df_final.columns.tolist())
```

# Crear columna derivada y guardado Final
 
La consigna pide crear la duracion en minutos

```python
# --- 7. Crear la columna derivada duracion_viaje_minutos ---
# Se calcula la diferencia entre la fecha/hora de destino y la de origen
df_final['duracion_viaje'] = (df_final['fecha_destino_recorrido'] - df_final['fecha_origen_recorrido'])

# Convertir el timedelta resultante a minutos (dividiendo por nanosegundos en un minuto)
MINUTO_EN_NANOSEGUNDOS = 60 * 10**9
df_final['duracion_viaje_minutos'] = df_final['duracion_viaje'].dt.total_seconds() / 60

# Opcional: eliminar la columna intermedia 'duracion_viaje' si no se necesita
df_final = df_final.drop(columns=['duracion_viaje'])

# Mostrar las primeras filas y estadísticas para verificar
print("\nPrimeras filas y duración del viaje en minutos:")
print(df_final[['fecha_origen_recorrido', 'fecha_destino_recorrido', 'duracion_viaje_minutos']].head())
print(df_final['duracion_viaje_minutos'].describe())
```

```python
# --- 8. Guardar el resultado procesado final ---
PROCESSED_PATH = '../data/processed/'
df_final.to_csv(PROCESSED_PATH + 'recorridos_usuarios_3meses.csv', index=False)

print(f"\nGuardado FINAL exitoso en {PROCESSED_PATH}recorridos_usuarios_3meses.csv")
```
